# **INTRODUCTION**

**Question**

The Kenyan Council of Governors is consulting on matters health and need insight on whether their counties meet the recommended doctor - population ratio of 1:1000 and if not, how they can go about achieving it.

They also require information on the distribution of health facilities in their counties.

**Research Objectives**
1. Do spatial visualization of the total population in Kenya
2. Show the distribution of health facilities in Kenya by counties
3. Map the accessibility of health facilities
4. Identify the proportion of health practitioners to county population
5. Show the different levels of healthcare in the county
6. Identify the number of health facilities in the counties that are government owned versus the number that are privately owned.


**Datasets Provided**

1. No of staff cadres per counties 2019 [https://test.hiskenya.org/kenya/dhis-web-dashboard-integration/index.action]
2. 2019-population_census-report-per-county [https://data.humdata.org/dataset/kenya-population-per-county-from-census-report-2019]
3. MFL_geocodes National [https://test.hiskenya.org/kenya/dhis-web-dashboard-integration/index.action]

# **Cleaning our Dataset**

### Merging our Datasets

```
# put the column names to lower case
facilities.columns = facilities.columns.str.lower()

pop.columns = pop.columns.str.lower()

officers.columns = officers.columns.str.lower()
```

```
# merge datasets

medics = facilities.merge(pop, how ='left',left_on ='fac_county', right_on = 'county').merge(officers, how = 'left',left_on = 'county',right_on='counties')
medics.tail()
```

### Question 6

```
# Question 6: Identify the number of health facilities in the counties that are government owned 
# versus the number that are privately owned.

# Find the number of facilities that are government owned per county
owner=medics[['fac_county','fac_owner']]
private = owner[~owner['fac_owner'].isin(['Ministry of Health'])]
government = owner[owner['fac_owner'].isin(['Ministry of Health'])]

# Government Facilities per county
gov_owned = government['fac_owner'].groupby(government['fac_county']).count()
gov_owned = pd.DataFrame(gov_owned)
gov_owned.columns = ['fac_gov_owned']

# Non-government Facilities per county
pr_owned = private['fac_owner'].groupby(private['fac_county']).count()
pr_owned = pd.DataFrame(pr_owned)
pr_owned.columns = ['fac_non_gov_owned']

# Total number of facilitities per county
total_owner = owner['fac_owner'].groupby(owner['fac_county']).count()

table = pd.concat([gov_owned, pr_owned,total_owner],axis=1)
table
```

```
# Number of Private and Government owned Facilities per County
owner['fac_owner'].value_counts()
```

```
# The ownership of facilities per County
owner['fac_owner'].groupby(owner['fac_county']).value_counts()
```
